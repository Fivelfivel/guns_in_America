```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
```

```python
df = pd.read_csv('cleandataset.csv', header=0)
df.head()
```

```python
def sentanal(tokens):
    sents=[]
    for tok in tokens:
        blob=TextBlob(tok, analyzer=NaiveBayesAnalyzer())
        sents.append(blob.sentiment)
    return sents
```

```python
s1=[]
for cont in df['content']:
    myl=cont.split('.')
    line1=myl[0]
    s1.append(line1)
```

```python
mysent=sentanal(s1)
```

```python
mysent[:5]
```

```python
import csv
```

```python
sentlist=[]
poslist=[]
neglist=[]
for item in mysent:
    sentlist.append(item.classification)
    poslist.append(item.p_pos)
    neglist.append(item.p_neg)
```

```python
df['sent']=sentlist
df['p_pos']=poslist
df['p_neg']=neglist
```

```python
df.head(5)
```

```python
df=df.drop('content', axis=1)
```

```python
df=df.drop('clean_content', axis=1)
```

```python
df.head(5)
```

```python
outfile = open('sentiments.csv', 'w')
df.to_csv(outfile)
```

```python
grp=df.groupby('paper').mean()
```
